# Create Schema and Tables
Run the SQL below to create the schema and volumes for the Sim Retail Demo. Be sure to select the desired catalog in the parameter before running.

Tables will be created in the Pyspark Bronze/Silver/Gold Notebooks and the SQL LDP file.

```sql
-- load catalog from parameter
USE CATALOG IDENTIFIER(:catalog);

-- create and use sim_retail_demo schema
CREATE SCHEMA IF NOT EXISTS sim_retail_demo;
USE SCHEMA sim_retail_demo;

-- create volumes to load raw JSON files
CREATE VOLUME IF NOT EXISTS customer;
CREATE VOLUME IF NOT EXISTS orders;
CREATE VOLUME IF NOT EXISTS status;
```
